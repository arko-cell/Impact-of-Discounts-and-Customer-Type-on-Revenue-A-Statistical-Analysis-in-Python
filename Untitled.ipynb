```python
import numpy as np
import pandas as pd
import os
import matplotlib as plt
import seaborn as sns
import statistics
```

```python
data_path = "C:/Users/ASUS/Documents/Project on marketing"   
```

```python
customers = pd.read_csv(
    os.path.join(data_path, "olist_customers_dataset.csv"),
    usecols=["customer_id", "customer_unique_id"]
)

orders = pd.read_csv(
    os.path.join(data_path, "olist_orders_dataset.csv"),
    usecols=["order_id", "customer_id", "order_purchase_timestamp", "order_status"]
)

payments = pd.read_csv(
    os.path.join(data_path, "olist_order_payments_dataset.csv"),
    usecols=["order_id", "payment_value"]
)

order_items = pd.read_csv(
    os.path.join(data_path, "olist_order_items_dataset.csv"),
    usecols=["order_id", "price", "freight_value"]
)
```

```python
df = pd.merge(
    orders,
    customers,
    on="customer_id",
    how="inner")

df = pd.merge(
    df,
    payments,
    on="order_id",
    how="inner")

df = pd.merge(
    df,
    order_items,
    on="order_id",
    how="inner")
```

```python
df["order_purchase_timestamp"] = pd.to_datetime(df["order_purchase_timestamp"])
```

```python
final_df = df.groupby(
    ["customer_unique_id", "order_id", "order_purchase_timestamp"],
    as_index=False
).agg({
    "payment_value": "sum",
    "price": "sum",
    "freight_value": "sum"
})
```

```python
final_df["revenue"] = final_df["payment_value"]-final_df["price"]
```

### Should the company prioritize customer retention over customer acquisition in its marketing strategy?

```python
customer_summary = final_df.groupby("customer_unique_id").agg(
    total_orders=("order_id", "nunique"),
    total_revenue=("revenue", "sum"),
    avg_order_value=("payment_value", "mean")
).reset_index()
```

```python
customer_summary.head(4)
```

```python
customer_summary["customer_type"] = customer_summary["total_orders"].apply(
    lambda x: "Repeat" if x > 1 else "One-Time"
)
```

```python
customer_summary.head(4)
```

```python
customer_summary.groupby("customer_type")["total_revenue"].describe()
```

```python
sns.boxplot(customer_summary['total_revenue'])
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt

stats.probplot(customer_summary['total_revenue'], plot=plt)
plt.show()
```

```python
from scipy.stats import skew, kurtosis

skewness = skew(customer_summary['total_revenue'])
kurt = kurtosis(customer_summary['total_revenue'])

skewness, kurt
```

```python

sns.histplot(customer_summary['total_revenue'].sample(5000), kde = True)
plt.show()
```

```python
from scipy.stats import ttest_ind

repeat_rev = customer_summary[
    customer_summary["customer_type"] == "Repeat"
]["total_revenue"]

one_time_rev = customer_summary[
    customer_summary["customer_type"] == "One-Time"
]["total_revenue"]

ttest_ind(repeat_rev, one_time_rev, equal_var=False)

```

```python
alpha = 0.05
p_value = 7.865516396136114e-13
if p_value < alpha:
    print ("accept null hypothesis")
else:
    print("Reject null Hypothesis")
```

```python
import numpy as np
from scipy.stats import sem, t

diff_mean = repeat_rev.mean() - one_time_rev.mean()

se = np.sqrt(sem(repeat_rev)**2 + sem(one_time_rev)**2)
ci = t.interval(0.95,
    df=len(repeat_rev)+len(one_time_rev)-2,
    loc=diff_mean,
    scale=se
               )

ci
```

Statistical testing confirms that repeat customers generate significantly higher revenue than one-time buyers (p < 0.05). The 95% confidence interval indicates a substantial profit difference. Therefore, reallocating marketing budget toward customer retention programs is statistically justified.

### Do discounted orders generate significantly different revenue compared to non-discounted orders?

```python
final_df["discount_applied"] = final_df["payment_value"] < final_df["price"]
final_df.head(10)
```

```python
final_df["discount_applied"].info
```

```python
final_df["discount_applied"].value_counts()
```

```python
final_df.groupby("discount_applied")["revenue"].describe()
```

Discounted orders show lower average revenue compared to non-discounted orders

```python


rev_discount = final_df[final_df["discount_applied"]]["revenue"]
rev_no_discount = final_df[~final_df["discount_applied"]]["revenue"]

t_stat, p_value = ttest_ind(
    rev_discount,
    rev_no_discount,
    equal_var=False
)

print("t-statistic:", t_stat)
print("p-value:", p_value)
```

```python
alpha = 0.05

if p_value < alpha:
    print("Discount has a statistically significant impact on revenue")
else:
    print("No statistically significant impact of discount on revenue")
```

```python
import numpy as np
from scipy.stats import sem, t

mean_diff = rev_no_discount.mean() - rev_discount.mean()

se = np.sqrt(sem(rev_no_discount)**2 + sem(rev_discount)**2)

ci = t.interval(
    0.95,
    df=len(rev_no_discount)+len(rev_discount)-2,
    loc=mean_diff,
    scale=se
)

ci
```

We are 95% confident that non-discounted orders generate more revenue than discounted orders.
